```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def func(i):

    f = open(f"../data/{i}.txt", "r")
    strings = f.readlines()
    li =[[float(x) for x in string.split()] for string in strings]
    return li
```

```python
lis = func("nums")
print(len(lis),len(lis[0]))

li = []
for x in lis:
    li+=x
```

```python
fig,_ = plt.subplots()
plt.hist(li, bins=64, alpha=0.5, color=["lightsteelblue"]*1, histtype='stepfilled')
plt.xlabel("patch loss")
plt.ylabel("patch frequency")
fig.savefig('../graph/patch_all.eps',dpi=600,format='eps')
```
